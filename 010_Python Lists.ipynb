### <font color="lightblue">**Python Lists**

### <font color="lightblue">**List**

List ถูกใช้เพื่อเก็บหลายๆค่าในตัวแปรเดียว

Lists เป็น 1 ใน 4 data type แบบ biult-in ใน ไพธอน ที่ใช้ในการเก็บค่า หลายๆค่าในตัวแปรเดียว โดยอีก 3 แบบ คือ Tuple Set และ Dictionary

เราสามารถสร้าง Lists ด้วย "วงเล็บสี่เหลี่ยม [ ]"

```python
thislist = ["apple", "banana", "cherry"]
print(thislist)
```

### <font color="lightblue">**ค่าใน List**

ค่าใน List นั้น "เป็นระเบียบ" "เปลี่ยนแปลงได้" และ "สามารถใช้ค่าซ้ำได้"

ทุกๆค่าของ List มีการระบุเลขตำแหน่งชัดเจน (เรียกว่า ค่าดัชนี) โดยค่าแรกใน List จะมีค่าดัชนี [0] และเรียงต่อๆไปเรื่อยๆ จนครบทุกค่าใน List

1.   เป็นระเบียบ

การที่ Lists เป็นระเบียบ หมายความว่าค่าใน List มีการระบุลำดับ และไม่มีการเปลี่ยนอันดับเลย เมื่อเราเพิ่มค่าใหม่ เข้าไปใน List ค่านั้นจะไปต่อท้ายสุดของ List เสมอ

2.   เปลี่ยนแปลง

การที่ Lists เปลี่ยนแปลงได้ หมายความว่าเราสามารถเพิ่ม เปลี่ยน หรือโละค่าใน List ออก หลังจากที่ถูกสร้างขึ้นมาเรียบร้อยแล้วได้

3.   สามารถใช้ค่าซ้ำได้

เนื่องจาก List มีการระบุเลขตำแหน่งชัดเจน จึงสามารถมีค่าที่ซ้ำ หรือมีค่าที่เหมือนกันได้


```python
thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(thislist)
```

### <font color="lightblue">**ความยาวของ List**

เราสามารถหาได้ว่า มีค่าอยู่ใน List อยู่เท่าไร โดยใช้ ฟังก์ชัน len():

```python
thislist = ["apple", "banana", "cherry"]
print(len(thislist))
```

### <font color="lightblue">**List Data Types**

ค่าใน List เป็น data type อะไรก็ได้

```python
list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]
```

List สามารถเก็บ data type ที่ต่างกันใน List ได้

```python
list1 = ["abc", 34, True, 40, "male"]
```

### <font color="lightblue">**type()**

จากมุมมองในตัวไพธอนนั้น List ถูกจัดเป็นค่าหนึ่งที่มี data type ของตัวเองเป็น 'list'

```python
mylist = ["apple", "banana", "cherry"]
print(type(mylist))
```

### <font color="lightblue">**The list() Constructor**

เราสามารถใช้ list() Constructor เพื่อสร้าง List ใหม่ได้

```python
thislist = list(("apple", "banana", "cherry")) # note the double round-brackets
print(thislist)
```

### <font color="lightblue">**Python Collections (Arrays)**

ในภาษาไพธอน มีการเก็บค่าตัวแปรเป็น collections อยู่ 4 แบบ
* List คือ collection ที่ เป็นระเบียบ เปลี่ยนแปลงได้ และ สามารถใช้ค่าซ้ำได้
* Tuple คือ collection ที่ เป็นระเบียบ แต่เปลี่ยนแปลงไม่ได้ สามารถใช้ค่าซ้ำได้
* Set คือ collection ที่ ไม่เป็นระเบียบ เปลี่ยนแปลงไม่ได้ และ ไม่สามารถใช้ค่าซ้ำได้
* Dictionary คือ collection ที่ เป็นระเบียบ เปลี่ยนแปลงได้ แต่ไม่สามารถใช้ค่าซ้ำได้
